```python
subfem = pd.read_csv('/content/drive/MyDrive/TCC - UFMS/TCC - UFMS - final/Análise de Frases Predicativas/Sentenças/df sem duplicatas.csv')
subfem.head()
submasc = pd.read_csv('/content/drive/MyDrive/TCC - UFMS/TCC - UFMS - final/Análise de Frases Predicativas/Sentenças/df masculino sem duplicata.csv')
submasc.head()
```

```python
dfmasc = dfmasc.rename(columns={' Adjetivo': 'Adjetivo'})
```

```python
def categorize_phrases(sub_df, adj_df):
    results = []
    for _, row in sub_df.iterrows():
        frase = row['Frase']
        matched = False
        for _, adj_row in adj_df.iterrows():
            adjetivo = adj_row['Adjetivo']
            categoria = adj_row['Categoria']
            if adjetivo in frase:
                results.append({'Frase': frase, 'Categoria': categoria})
                matched = True
                break
        if not matched:
            results.append({'Frase': frase, 'Categoria': None})
    return pd.DataFrame(results)

fem_result_df = categorize_phrases(subfem, dffem)
masc_result_df = categorize_phrases(submasc, dfmasc)

fem_result_df.head()
masc_result_df.head()
```

```python
result_df = pd.concat([fem_result_df, masc_result_df], ignore_index=True)

result_df.head()
```
